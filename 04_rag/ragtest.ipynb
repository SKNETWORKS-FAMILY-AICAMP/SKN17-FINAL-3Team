```python
import os
import unicodedata
import re
from langchain_community.vectorstores import FAISS
from langchain_openai import OpenAIEmbeddings

DB_PATH = "./faiss_index"

def clean_text(text):
    """제목 비교를 위한 강력한 정제 함수"""
    if not text: return ""
    text = unicodedata.normalize('NFC', str(text))
    return re.sub(r"[^0-9a-zA-Z가-힣]", "", text)

def fetch_full_match_log():
    print("📂 DB 로딩 중...")
    
    if not os.path.exists(DB_PATH):
        print("❌ DB 폴더가 없습니다.")
        return

    embeddings = OpenAIEmbeddings()
    try:
        # DB 로드
        vector_db = FAISS.load_local(DB_PATH, embeddings, allow_dangerous_deserialization=True)
    except Exception as e:
        print(f"❌ DB 로드 실패: {e}")
        return

    # =======================================================
    # [설정] 가져오고 싶은 경기 제목
    # =======================================================
    target_match = "2025 한국시리즈 1차전"
    
    print(f"🎯 타겟 경기: '{target_match}' 의 전체 기록을 가져옵니다...")
    
    target_clean = clean_text(target_match)
    all_docs = vector_db.docstore._dict
    matched_chunks = []

    for doc_id, doc in all_docs.items():
        db_title = doc.metadata.get("match_title", "")
        
        if target_clean in clean_text(db_title):
            matched_chunks.append(doc.page_content)

    # =======================================================
    # [결과 처리] 조각 합치기
    # =======================================================
    if not matched_chunks:
        print("❌ 해당 경기의 데이터를 찾을 수 없습니다.")
        return

    print(f"\n✅ 총 {len(matched_chunks)}개의 조각을 발견했습니다. 하나로 합칩니다...\n")
    
    # 조각들을 줄바꿈으로 연결
    full_text = "\n\n".join(matched_chunks)
    
    print("=" * 50)
    print(f"   [ {target_match} 전체 기록 ]")
    print("=" * 50)
    print(full_text)
    print("=" * 50)
    
    return full_text

if __name__ == "__main__":
    full_data = fetch_full_match_log()
```
